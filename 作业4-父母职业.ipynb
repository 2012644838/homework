```python
import pandas as pd
import mytools
```

```python
列名列表 = ['父亲职业','父亲工作','母亲职业','母亲工作']
df1 = pd.read_csv(R'类别变量练习数据.csv', encoding='GBK', usecols=列名列表)
```

```python
import matplotlib.pyplot as plt
x=df1['母亲职业'].value_counts().index
y=df1['母亲职业'].value_counts(normalize=True).values
fig,ax2=plt.subplots()
labels=df1['母亲职业'].value_counts().index
ax2.pie(y,autopct='%1.1f%%',labels=labels,textprops=dict(color='w'))
plt.show
```

```python
cross_table = pd.crosstab(df1['父亲职业'], df1['母亲职业'])
print(cross_table)
```

```python
tau_y = mytools.goodmanKruska_tau_y(df1,'父亲职业','母亲职业')
tau_y
```
