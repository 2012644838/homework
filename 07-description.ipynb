```python
#单变量描述统计
##无序类别变量描述统计
import pandas as pd 
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #设置字体
df= pd.read_spss(R'identity.sav')
```

```python
x=df['政治面貌'].value_counts().index
y=df['政治面貌'].value_counts(normalize=True).values
fig,ax2=plt.subplots()
labels=df['政治面貌'].value_counts().index
ax2.pie(y,autopct='%1.1f%%',labels=labels,textprops=dict(color='w'))
plt.show
```

```python
import matplotlib.pyplot as plt

def plot_pie_chart(data, variable):
    
    counts = data[variable].value_counts()

    plt.figure(figsize=(8, 6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')  # 使饼图保持圆形
    plt.title(f' {variable}')
    plt.show()
```
